```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,10))
plt.hist(df['Weather_conditions'],bins=30)
plt.xlabel("Weather Conditions")
plt.title("Effect of weather condtiond on accidents")
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.hist(df['Accident_severity'],bins=10,color='pink',label='accident_seveirity')
plt.title("Severity of accidents")
plt.show()
```

```python

counts = df['Type_of_vehicle'].value_counts()
plt.figure(figsize=(20, 20))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140)

# Adding a title
plt.title('Distribution of Vehicle Types Involved in Traffic Accidents')

# Display the pie chart
plt.show()

```

```python
counts = df['Type_of_collision'].value_counts()
plt.figure(figsize=(20, 20))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140)

# Adding a title
plt.title('Distribution of collision Types Involved in Traffic Accidents')

# Display the pie chart
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting the count plot
plt.figure(figsize=(20, 6))
sns.countplot(x='Type_of_collision', data=df, palette='viridis')

# Adding a title and labels
plt.title('Count of Vehicle Types Involved in Traffic Accidents')
plt.xlabel('Vehicle Type')
plt.ylabel('Count')

# Display the plot
plt.show()
```
